## Scatter plot

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("countryGDP.csv")
```

```python
df.head()
```

```python
df1 = df.dropna()
```

```python
plt.scatter(df1['Population'], df1['GDP per Capita'],
           marker = 'x',
           color = 'b',
           alpha = .7,
            s = 124)

plt.title("National GDP vs Pop.")
plt.xlabel("Population")
plt.ylabel("GDP in Dollars")
plt.show()
```

```python
from sklearn import datasets
import seaborn as sns
```

```python
iris = datasets.load_iris()
iris = pd.DataFrame (data = np.c_[iris['data'],iris['target']],columns = iris['feature_names'] + ['species'])
iris.species = np.where (iris.species == 0.0, 'setosa', np.where (iris.species == 1.0, 'versicolor','virginica') )
iris.columns = iris.columns.str.replace(' ','')

iris.head (6)
```

```python
# CHECK
iris.plot(kind = 'scatter', x = 'sepallength(cm)', y = 'sepalwidth(cm)' )
```

```python
# CHECK
sns.FacetGrid(iris,hue='species',size=5).map(plt.scatter, 'sepallength(cm)', 'sepalwidth(cm)' ).add_legend()
```
