```python
#@matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
print (type(iris))
```

```python
import pandas as pd
```

```python
iris = pd.DataFrame(data = np.c_[iris['data'],iris['target']], columns = iris['feature_names']+['species'])
```

```python
iris.head(6)
```

```python
iris.species = np.where (iris.species == 0.0, 'setosa', np.where (iris.species == 1.0, 'versicolor','virginica') )
```

```python
iris.columns = iris.columns.str.replace (' ','')
```

```python
iris.head(5)
```

```python
iris.hist()
```

```python
df = pd.read_csv('CountryGDP.csv')
df.head(5)
```

```python
data1 = df['GDP per Capita']
data1.head(5)
```

```python
data1.hist()
```

```python
plt.figure();
df['GDP per Capita'].diff().hist()
```

```python
import numpy as np
```

```python
b = np.arange(data1.min(),data1.max(),1000)
print(b)
```

```python
df1 = data1.dropna()
```

```python
plt.hist(df1,
        bins = b,
        alpha = 0.5,
         color = 'green',
         label = 'GDP')
plt.title("National GDP's")
plt.xlabel("GDP in DOllars")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.hist(df1,bins = 20, histtype = 'step')
```

```python
import seaborn as sns
```

```python
plt.hist(df1,alpha = .3)
sns.rugplot(df1);
```

```python
tips = sns.load_dataset ("tips")
tips.head(4)
```

```python
g = sns.FacetGrid(tips,row= 'sex', col = 'time', margin_titles = True)
bins = np.linspace(0,60,13)
g.map(plt.hist,"total_bill",color= 'steelblue", bins=bins, lw=0)
```
