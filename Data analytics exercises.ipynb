## 1. Na jaké kategorii produktů máme největší obrat? A zajímalo by mě i jestli se to v jednotlivých měsících mění.

```python
data_Tran["Date"] = pd.to_datetime(data_Tran["Date"])
data_Tran.head()
```

```python
data_Tran ["Year-Month"] = data_Tran["Date"].dt.to_period ("M")
data_Tran.head()
```

```python
data_Tran = pd.merge(data_Tran,data_Prod, on = "Product name", how="left")
data_Tran.head()
```

```python
data_Tran["Sales"]=data_Tran["Price"]*data_Tran["Quantity"]
data_Tran.head()
```

```python
Grouped_month = data_Tran.groupby(["Product name","Year-Month"])["Sales"].sum().reset_index()
Grouped_month.head(7)
```

```python
Grouped = data_Tran.groupby("Product name")["Sales"].sum().reset_index()
Grouped.head(7)
```

```python
Grouped = Grouped.sort_values(by="Sales",ascending=False)
Grouped.head(7)
```

```python
Grouped_month = Grouped_month.sort_values(by="Sales",ascending=False)
Grouped_month.head(7)
```
