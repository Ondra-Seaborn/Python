```python
import pandas as pd
import numpy as np
```

```python
gfp = pd.read_csv("C:\\Users\\onvac\\OneDrive\\Desktop\\Python\\Training\\Udemy course\\MyTraining files\\ForAnaconda\\5_Data\\GlobalFirePower.csv")
gfp.head(6)
```

```python
gdp = pd.read_csv("C:\\Users\\onvac\\OneDrive\\Desktop\\Python\\Training\\Udemy course\\MyTraining files\\ForAnaconda\\5_Data\\CountryGDP.csv")
gdp
```

```python
x = pd.merge(gdp,gfp,on='Country')
#inner is default
x.head()
```

```python
y = pd.merge(gdp,gfp,on='Country',how='left')
# how = left,right,outer
y.head()
```

```python
y = pd.merge(gdp,gfp,on='Country',how='outer',indicator = True)
y.head()
```
